### What is a Generator
Python generators are a simple way of creating iterators.

```python
def gen_demo():
    yield 'first statement' 
    yield 'second statement' 
    yield 'third statement'
```

```python
gen = gen_demo()
# print(next(gen))
# print(next(gen))
# print(next(gen))
for i in gen: 
    print(i)
```

### yield vs return 

```python
def square(num):
    for i in range(1,num+1): 
        yield i**2
```

```python
gen = square(10)
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
print('------------------')
for i in gen: 
    print(i)
```

### range function using generator 

```python
def my_range(start, end): 
    for i in range(start,end): 
        yield i 
```

```python
gen = my_range(15,26)
for i in gen: 
    print(i)
```

### generator expression 

```python
# list comprension  
l = [i**2 for i in range(1,15)] 
```

```python
gen = (i**2 for i in range(1,15))
for i in gen: 
    print(i)
```
