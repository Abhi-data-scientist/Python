# Abstraction


```python
# abstract class ->  jisme koi abstract method hota hai
# inhertited class must use something which parent wants
from abc import ABC, abstractmethod
class BankApp(ABC):
    
    def database(self):
        print('connected to database')

    @abstractmethod
    def security(self):
        pass
```

```python
class mobileApp(BankApp):
    def mobile_login(self): 
        print('Login into mobile')

    def security(self): 
        print('Mobile security')
```

```python
mob = mobileApp()
```

```python
mob.database()
```

```python
mob.security()
```
