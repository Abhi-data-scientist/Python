```python
# aggregation
class Customer:
  def __init__(self, name, gender, address):
    self.name = name
    self.gender = gender
    self.address = address

  def print_address(self):
    print(self.address.get_city(), self.address.pin, self.address.state)

  def edit_profile(self, new_name, new_city, new_pin, new_state):
    self.name = new_name
    self.address.edit_address(new_city, new_pin, new_state)


class Address:
  def __init__(self, city, pin, state):
    self.__city = city
    self.pin = pin
    self.state = state

  def get_city(self):
    return self.__city

  def edit_address(self, new_city, new_pin, new_state):
    self.__city = new_city
    self.pin = new_pin
    self.state = new_state


add1 = Address('jaipur', 302029, 'rajasthan')
cust = Customer('Abhishek', 'male', add1)
# cust.print_address()

cust.edit_profile('Abhi', 'californiya', 53435, 'America')
cust.print_address()
```

```python
# Parent
class User:
  def __init__(self):
    self.name = "Abhi"

  def login(self):
    print("login")

# child
class Student(User):
  def enroll(self):
    print('Enroll into the course')

u = User()
s = Student()

print(s.name)
s.login()
s.enroll()
```

***Super keyword***
* by using this we can access the methods of parent class
* class ke bahar super keyword ko use nhi karte
* only child class me use hota hai


```python
class phone :
  def __init__(self, price, brand, camera):
    print("Inside phone constructor")
    self.price = price
    self.brand = brand
    self.camera = camera

class SmartPhone(phone):
  def __init__(self, price, brand, camera, os, ram):
    print("inside smartphone constructor")
    super().__init__(price, brand, camera)
    self.os = os
    self.ram = ram
    print("inside smartphone constructor")

s = SmartPhone(2000,'samsung', 14, 'android', 4)
print(s.os)
print(s.price)
```

***Inheritance***
* A class can inherit from another class
* inheritance improves code reuse
* constructor, attributes, methods, get inherited to the child class
* the parent has no access to the child class
* private properties of parent are not accessible directly in child class
* child class can override the attributes or methods this is called overriding
* super() is an inbuilt function which is used to invoke the parent class methods and constructor

***Types of inheritance***
* singal inheritance 
* multilevel inhertance
* hierarchial inheritance
* multiple inheritance 
* hybrid inheritance

```python
# 1. singal inheritance
class Phone :
  def __init__(self, price, brand, camera):
    print('Inside phone constructor')
    self.price = price 
    self.brand = brand
    self.camera = camera

  def buy(self):
    print('Buying a phone')

class SmartPhone(Phone):
  pass 

SmartPhone(1000, 'Iphone', '34px').buy()
```

```python
# 2. Multilevel inheritance 
class Product:
  def review(self):
    print('Product customer review')

class Phone(Product):
  def __init__(self, price, brand, camera):
    print('Inside phone constructor')
    self.price = price 
    self.brand = brand
    self.camera = camera

  def buy(self):
    print('Buying a phone')

class SmartPhone(Phone):
  pass 

s = SmartPhone(2000, 'Apple', 15)
s.buy()
s.review()
```

```python
# 3. Hierarchial inheritance 
class Phone():
  def __init__(self, price, brand, camera):
    print('Inside phone constructor')
    self.price = price 
    self.brand = brand
    self.camera = camera
  def buy(self):
    print('Buying a phone')
  
class SmartPhone(Phone):
  pass 

class KeyPhone(Phone):
  pass

s = SmartPhone(2000,'Apple', 12).buy()
k = KeyPhone(10, 'lava', 1).buy()
```

```python
# Multiple 
class Phone():
  def __init__(self, price, brand, camera):
    print('Inside phone constructor')
    self.price = price 
    self.brand = brand
    self.camera = camera
  def buy(self):
    print('Buying a phone')
  
class Product:
  def review(self):
    print('Product customer review')

class SmartPhone(Phone, Product):
  pass 
s = SmartPhone(20000, 'Apple', 12)
s.buy()
s.review()
```

```python
# Method resolution order
class Phone():
  def __init__(self, price, brand, camera):
    print('Inside phone constructor')
    self.price = price 
    self.brand = brand
    self.camera = camera
  def buy(self):
    print('Buying a phone')
  
class Product:
  def buy(self):
    print('Product buy method')

class SmartPhone(Phone, Product): # jo phle likhte hai uska method call hota hai 
  pass 
s = SmartPhone(20000, 'Apple', 12)
s.buy()
```
