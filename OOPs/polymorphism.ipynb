# Method overloading 
when we have same name methods but both have different output based on their inputs
it does not work in python 

why we need =>
    because of good reading
    code looks clean

```python
class shape:
    def area(self, radius):
        return 3.14*radius*radius 

    def area(self, l, b):
        return l*b;

s = shape()
s.area(2)
s.area(3,4)
```

```python
class shape:
    def area(self, a, b=0): # b is default keyword
        if b==0:
            return 3.14*a*a
        else:
            return a*b 

s = shape()
print(s.area(2))
print(s.area(3,4)) 
```
