### Namespaces

A namespace is a space that holds names(identifiers).Programmatically speaking, namespaces are dictionary of identifiers(keys) and their objects(values)

There are 4 types of namespaces:
- Builtin Namespace
- Global Namespace
- Enclosing Namespace
- Local Namespace

### Scope and LEGB Rule

A scope is a textual region of a Python program where a namespace is directly accessible.

The interpreter searches for a name from the inside out, looking in the local, enclosing, global, and finally the built-in scope. If the interpreter doesnâ€™t find the name in any of these locations, then Python raises a NameError exception.

```python
# local and global 
a = 2  # global scope 
def temp():
    b = 3  # local scope
    print(b)
temp()
a
```

```python
# local and global same name 
a = 2  # global scope 
def temp():
    a = 3  # local scope
    print(a)
temp()
a
```

```python
# local and global -> local does not have but global has 
a = 2  # global scope 
def temp():
    print(a)
temp()
a
```

```python
# local and global -> edditing global variable 
#                -> we can view global variable but can not change them 
a = 2  # global scope 
def temp():
    a+=1
    print(a)
temp()
a
```

```python
# global keyword se change kr skte hai 
a = 2  # global scope 
def temp():
    global a
    a+=1
    print(a)
temp()
a
```

```python
# local and global -> global created inside local 
def temp():
    global a
    a=1
    print(a)
temp()
a
```

```python
# local and global -> function parameter is local 
def temp(z): 
    print(z) 
a = 5
temp(5)
print(a)
```

```python
 # Enclosing scope  -> nested function me milta hai 
# global
def outer(): 
    # enclosing scope 
    def inner(): 
        # local scope
        print('inner function')
    inner()
    print('outer function') 

outer()
print('main function')
```

```python
 # global
def outer(): 
    a=3
    # enclosing scope 
    def inner(): 
        a=4
        # local scope
        print(a)
    inner()
    print('outer function') 

outer()
print('main function')
```

```python
 # global
def outer(): 
    # enclosing scope / nonlocal scope
    a=3
    def inner(): 
        # local scope
        print(a)
    inner()
    print('outer function') 

outer()
print('main function')
```

```python
 # global
def outer(): 
    # enclosing scope 
    a=3
    def inner(): 
        # local scope
        nonlocal a 
        a+=1
        print('inner ', a)
    inner()
    print('outer ', a) 

outer()
print('main function')
```
